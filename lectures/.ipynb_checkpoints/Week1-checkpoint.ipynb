```python
import pandas as pd
df = pd.read_csv('../files/Police_Department_Incident_Reports__2018_to_Present.csv')
df.head()
```

```python
#Exercise 3
import matplotlib.pyplot as plt

crimes = df['Incident Year']
#crimes


dict = {}

def findYear():
    for i in range (len(crimes)):
        if crimes[i] in dict.keys():
            dict[crimes[i]] += 1
        else:
            dict[crimes[i]] = 1  
    return dict
        
findYear()



plt.figure(figsize=(10, 3), dpi=80)
plt.xlabel("Year")
plt.ylabel("Number of crimes")
plt.bar(dict.keys(), dict.values(), color = 'blue', label = 'Neutral')


    
```

```python
#Exercise 3:
values = dict.values()
total_number_crimes = sum(values)


average = total_number_crimes/len(dict)
print("The average amount of crimes each year is: " + str(average))

```

```python
#Exercise 1:

total_crimes = len(crimes)
print("Total number of crimes is:" + str(total_crimes))
```

```python
#Exercise 1:
categories = df['Incident Category']
categories


dict_categories = {}

def findCategories():
    for i in range (len(crimes)):
        if categories[i] in dict_categories.keys():
            dict_categories[categories[i]] += 1
        else:
            dict_categories[categories[i]] = 1  
    return dict_categories

print(findCategories())

print("Total number of categories is: " + str(len(dict_categories)))


```

```python
newlist = []
for i in dict_categories.keys():
    if (type(i)) != float:
        newlist.append(i)
        
print(newlist)
    
#Problem: one of the keys is a float

plt.figure(figsize=(10, 3), dpi=80)
plt.xlabel("Category")
plt.ylabel("Number of crimes")


plt.bar(newlist, dict_categories.values(), color = 'blue', label = 'Neutral')
```

```python
#Exercise 2

max_number_of_crimes = max(dict_categories.values())

#max_category = dict_categories.keys()[dict_categories.values().index(max_number_of_crimes)]


for key, value in dict_categories.items():
     if max_number_of_crimes == value:
            print("The category with most crimes is: " + str(key))
            
min_number_of_crimes = min(dict_categories.values())

for key, value in dict_categories.items():
     if min_number_of_crimes == value:
            print("The category with least number of crimes is: " + str(key))
            
```

```python
#Exercise 3:

#the average number of crimes per year by category


df_year = df.groupby(['Incident Category'])['Incident Year'].value_counts()

#print(df_year)
#print(df_year['Arson'])

  

def get_plot(category):
    dict_year_category = {}
    for i in range(len(crimes)):
        cat = str(df['Incident Category'][i]).upper()
        if cat == category:
            year = df['Incident Year'][i]
            if year in dict_year_category:
                dict_year_category[year] += 1
            else:
                dict_year_category[year] = 0
    plt.figure(figsize=(10, 3), dpi=80)
    plt.title(category)
    plt.xlabel("Year")
    plt.ylabel("Number of crimes")
    plt.bar(dict_year_category.keys(), dict_year_category.values(), color = 'blue', label = category)
    

    
    
get_plot('BURGLARY')
        


```

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

```python
for i in focuscrimes:
    get_plot(i)
```
